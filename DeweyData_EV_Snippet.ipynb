```python
from dotenv import load_dotenv
import googlemaps
import os

load_dotenv()
api_key = os.getenv('GOOGLE_MAPS_API_KEY')
gmaps = googlemaps.Client(key=api_key)

locations = df[['STREET_ADDRESS', 'CITY', 'REGION', 'POSTAL_CODE']].drop_duplicates().dropna()

locations['FULL_ADDRESS'] = locations['STREET_ADDRESS'] + ', ' + locations['CITY'] + ', ' + locations['REGION'] + ', ' + locations['POSTAL_CODE'].astype(str)
```

```python
locations
```

```python
true_locations = None

for index, row in locations.iterrows():
    full_address = row['FULL_ADDRESS']
    try:
        geocode_result = gmaps.geocode(full_address)
        place_id = geocode_result[0]['place_id']
        place_details = gmaps.place(place_id)

        business_name = place_details['result'].get('name', 'Unknown')
        latitude = place_details['result']['geometry']['location']['lat']
        longitude = place_details['result']['geometry']['location']['lng']
        formatted_address = place_details['result'].get('formatted_address', 'Unknown')
        location_type = place_details['result'].get('types', ['Unknown'])[0]
        location_id = place_details['result'].get('place_id', 'Unknown')
        location = {
            'business_name': business_name,
            'latitude': latitude,
            'longitude': longitude,
            'formatted_address': formatted_address,
            'location_type': location_type,
            'location_id': location_id
        }
        location_df = pd.DataFrame([location])
        if true_locations is None:
            true_locations = location_df
        else:
            true_locations = pd.concat([true_locations, location_df], ignore_index=True)
    except Exception as e:
        print(f"Error processing address {full_address}: {e}")

# Save the true locations to a CSV file
true_locations.to_csv('true_locations.csv', index=False)
```
